### Reading and Displaying an Image

```python
path = 'basic_concepts/clock_1.jpeg'
original = cv2.imread(path,1) # Read it as color image
print(original.shape)


ImgColor=cv2.cvtColor(original, cv2.COLOR_BGR2RGB) 

#Imgcolor=cv2.cvtColor(original, )
# gray scale
ImgGray=cv2.imread(path,0)# Read it as gray
plt.imshow(ImgColor[:]/255.0)
plt.axis('off')
plt.show()
```

```python
#Know the Image Dimension
print('The color image is of dimension:',ImgColor.shape,'type:',type(ImgColor))
```

```python
h,w,c = ImgColor.shape
print('Image is of Height {0},Width {1} with {2} channels '.format(h,w,c))
```

```python
print(ImgGray.shape)
print(ImgColor.ndim)
```

```python
#Access Pixel Values
pixel=ImgColor[0,0] 
print(pixel)
```

```python
# Extract Three channels, It is a costly operation
r,g,b=cv2.split(ImgColor)  
plt.figure(num=None,figsize=(10,8),dpi=80)
plt.subplot(131)
plt.imshow(b[:]/255.0,cmap='gray')
plt.title('Blue')
plt.axis('off')
plt.subplot(132)
plt.imshow(g[:]/255.0,cmap='gray')
plt.title('Green')
plt.axis('off')
plt.subplot(133)
plt.imshow(r[:]/255.0,cmap='gray')
plt.title('Red')
plt.axis('off')
plt.show()
```

```python
# Making Combination of channels
Img=ImgColor.copy() 
Img[:,:,1] = 0 
plt.imshow(Img[:]/255.0)
plt.title('Combination')
plt.axis('off')
plt.show()
```

```python
# Inverting
r,g,b=cv2.split(ImgColor) 
R = 255-b 
b[:],r[:] = 0,0
blue_invert= cv2.merge((b,g,R))
plt.imshow(blue_invert[:]/255.0)
plt.title('Red inversion')
plt.axis('off')
plt.show()
print('The color image is of dimension:',ImgColor.shape,'type:',type(ImgColor))
```

```python
# Making Dominant channels
Img=ImgColor.copy() # if copy() not used changing Img affects ImgColor too (Original)
Img[:,:,(0,1)]=0 
plt.imshow(Img[:]/255.0)
plt.title('Combination')
plt.axis('off')
plt.show()
```

### Intensity Transformation

```python
ret,thresh1 = cv2.threshold(ImgGray,100,255,cv2.THRESH_BINARY) # Step change
ret,thresh2 = cv2.threshold(ImgGray,127,255,cv2.THRESH_BINARY_INV)#Inverse of above
ret,thresh3 = cv2.threshold(ImgGray,170,255,cv2.THRESH_TRUNC)   
ret,thresh4 = cv2.threshold(ImgGray,200,255,cv2.THRESH_TOZERO) # Keep value above thrsh as it is, make others zero (ramp)
ret,thresh5 = cv2.threshold(ImgGray,127,255,cv2.THRESH_TOZERO_INV)#Inverse of the above
```

```python
plt.figure(num=None,figsize=(10,8),dpi=80)
plt.subplot(121)
plt.imshow(ImgGray[:]/255.0,cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(122)
plt.imshow(thresh1[:]/255.0,cmap='gray')
plt.title('Thresholded1')
plt.axis('off')
plt.show()
plt.imshow(thresh2[:]/255.0,cmap='gray')
plt.title('Thresholded2')
plt.axis('off')
plt.show()
plt.imshow(thresh3[:]/255.0,cmap='gray')
plt.title('Thresholded3')
plt.axis('off')
plt.show()
plt.imshow(thresh4[:]/255.0,cmap='gray')
plt.title('Thresholded4')
plt.axis('off')
plt.show()
plt.imshow(thresh5[:]/255.0,cmap='gray')
plt.title('Thresholded5')
plt.axis('off')
plt.show()
```
